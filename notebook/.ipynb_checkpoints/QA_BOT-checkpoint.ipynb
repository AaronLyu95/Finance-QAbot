# Document Splitting

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1024, 
    chunk_overlap=64, 
    separators=["\n\n", "\n", "(?<=\. )", " ", ""])
```

```python
texts = text_splitter.split_documents(pages)
len(texts)
```
